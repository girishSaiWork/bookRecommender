```python
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
plt.figure(figsize = (8,6))
plt.style.use('fivethirtyeight')
ax = plt.axes()
sns.heatmap(books.isna().transpose(), cbar=False, ax=ax)

plt.xlabel("Columns")
plt.ylabel("Missing values")

plt.show()
```

```python
current_year = datetime.now().year
books['missing_description'] = np.where(books['description'].isna(), 1,0)
books['age_of_book'] = current_year - books['published_year']
```

```python
interested_columns = ['num_pages','missing_description','age_of_book','average_rating']
p_correlation_matrix = books[interested_columns].corr(method = 'pearson')
correlation_matrix = books[interested_columns].corr(method = 'spearman')
plt.figure(figsize = (13,8))

# Main title for the entire figure
plt.suptitle("Comparison of Spearman and Pearson Correlation Matrices", fontsize=16)  # Adjust fontsize as needed

plt.subplot(1,2,1)
sns.set_theme(style="white")
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f',
            cbar_kws={'label': 'Spearman Correlation'})
plt.title("Spearman Correlation") # Subplot title

plt.subplot(1,2,2)
sns.set_theme(style="white")
hm2 = sns.heatmap(p_correlation_matrix, annot=True, cmap='Blues', fmt='.2f',
            cbar_kws={'label': 'Pearson Correlation'})
plt.title("Pearson Correlation") # Subplot title


plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Adjust layout to prevent overlap with suptitle
plt.show()
```

```python
book_missing = books[~(books["description"].isna()) &
      ~(books["num_pages"].isna()) &
      ~(books["average_rating"].isna()) &
      ~(books["published_year"].isna())
]
book_missing
```

```python
book_missing["words_in_description"] = book_missing['description'].str.split().str.len()
book_missing
```

```python
book_missing_25_words =book_missing.loc[book_missing["words_in_description"] >= 25]
book_missing_25_words
```

```python
book_missing_25_words["title_and_subtitle"] = (
    np.where(book_missing_25_words["subtitle"].isna(), book_missing_25_words["title"],
             book_missing_25_words[["title", "subtitle"]].astype(str).agg(": ".join, axis=1))
)
book_missing_25_words
```

```python
book_missing_25_words["tagged_description"] = book_missing_25_words[["isbn10","description"]].astype(str).agg(" ".join, axis=1)
book_missing_25_words
```

```python
(
    book_missing_25_words
    .drop(["subtitle", "missing_description", "age_of_book", "words_in_description"], axis=1)
    .to_csv("books_cleaned.csv", index = False)
)
```

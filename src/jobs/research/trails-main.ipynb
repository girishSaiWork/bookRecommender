```python
import os
import pandas as pd
import numpy as np

from langchain_ollama import OllamaEmbeddings
from langchain_community.vectorstores import PGVector
# from langchain_postgres.vectorstores import PGVector
import warnings
warnings.filterwarnings('ignore')
```

```python
filepath = r"C:\Users\saigi\PycharmProjects\bookRecommender\data\books_with_emotions.csv"
cover_image_path = r"C:\Users\saigi\PycharmProjects\bookRecommender\data\cover-not-found.jpg"
print("Reading data file...")
books = pd.read_csv(filepath)
books.head()
```

```python
books["large_thumbnail"] = books["thumbnail"] + "&fife=w800"
books["large_thumbnail"] = np.where(
    books["large_thumbnail"].isna(),
    cover_image_path,
    books["large_thumbnail"],
)

books.head()
```

```python
embedding_function = OllamaEmbeddings(model="mxbai-embed-large")
```

```python
CONNECTION_STRING = r"postgresql://postgres:lang@localhost:5432/vectordb"
COLLECTION_NAME = 'books_description_vectors'
```

```python
# Initialize PGVector using the existing data in PostgreSQL
db_books = PGVector(
    connection_string=CONNECTION_STRING,
    collection_name=COLLECTION_NAME,
    embedding_function=OllamaEmbeddings(model="mxbai-embed-large"),
    use_jsonb=True,
)
```

```python
query = "A book to teach children about nature"
docs = db_books.similarity_search(query, k = 10)
docs
```
